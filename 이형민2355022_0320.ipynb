```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
url = "https://raw.githubusercontent.com/ancestor9/data/main/customers.csv"
```

```python
df = pd.read_csv(url)
df
```

```python
summary = pd.DataFrame({
    'Data Type': df.dtypes,
    'Unique Values': df.nunique(),
    'Missing Values': df.isnull().sum(),
    'Missing Percentage (%)': (df.isnull().sum() / len(df)) * 100
})

print("--- [데이터 요약 정보] ---")
print(summary)
```

```python
try:
    df = pd.read_csv(url)
    print("데이터 로드 성공!")
except Exception as e:
    print(f"데이터 로드 실패: {e}")
    # 실패 시 샘플 데이터 생성 (구조 파악용)
    data = {
        'CustomerID': [1, 2, 3],
        'Gender': ['Male', 'Female', 'Female'],
        'Age': [19, 21, 20],
        'Annual Income (k$)': [15, 15, 16],
        'Spending Score (1-100)': [39, 81, 6]
    }
    df = pd.DataFrame(data)
```

```python
if 'CustomerID' in df.columns:
    df = df.drop('CustomerID', axis=1)
```

```python
print(df.columns)
```

```python
le = LabelEncoder()
df['성별'] = le.fit_transform(df['성별'])
```

```python
df_encoded = pd.get_dummies(df, columns=['성별'], drop_first=False)
```
